# Manipulação de dados com polars

```python
import polars as pl
import numpy as np
from matplotlib import pyplot as plt
```

```python
pl.__version__
```

```python
x = np.arange(0.001, 100000.001, 0.001)
y = [(1 + 1/i)**i for i in x]
```

```python
max(y)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y, label='e')

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_xlim(0, 10)

plt.show()
```
